```python
%%capture

from txtai.pipeline import Translation 

# Create translation model
translate = Translation()
```

```python
translation = translate("Love is no love", "zh")
translation
```

```python
languages = ["fr", "en", "zh", "hi", "ja"]

translations = [translate("Language Translation is Excellent", language) for language in languages]

english = translate(translations, "en")

for x, text in enumerate(translations):
  print("Original langues: %s" % languages[x])
  print("Translation: %s" % text)
  print("Back to English: %s" % english[x])
  print()
```
